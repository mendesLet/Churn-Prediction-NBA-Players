```python
import matplotlib.pyplot as plt

models = ['Medium', 'FMF']
accuracy = [0.84, 0.84]  # Valores de acurácia
precision = [0.84, 0.84]  # Valores de precisão
f1_score = [0.84, 0.83]  # Valores de F1-score
recall = [0.83, 0.82]  # Valores de recall

plt.figure(figsize=(10, 6))

plt.plot(models, accuracy, marker='o', label='Acurácia')
plt.plot(models, precision, marker='o', label='Precisão')
plt.plot(models, f1_score, marker='o', label='F1-score')
plt.plot(models, recall, marker='o', label='Recall')

plt.title('Comparação de Métricas entre Modelos de Logistic Regression:')
plt.xlabel('Modelos')
plt.ylabel('Métricas')
plt.legend()
plt.grid(True)
plt.show()
```
